```python
import csv
 
def csv_writer(data, path):
    """
    Write data to a CSV file path
    """
    with open(path, "w", newline='') as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        for line in data:
            writer.writerow(line)
 
if __name__ == "__main__":
    data = ["Mall, Far from metro, Number of transport stops, Number of parking lots, Number of building apartment complexes".split(","),
            "Aviapark,450,2,5000,5".split(","),
            "Kashirskaya plaza,1200,2,1500,2".split(","),
            "Mega Khimki,,1,8500,3".split(","),
            "Metropolis,350,3,2820,4".split(","),
            "Riviera,1300,3,3000,2".split(",")
            ]
    
    path = "output.csv"
    csv_writer(data, path)
```

### Средняя посещаемость ТЦ на выходных

```python
sunday = mall_data[mall_data['Weekday'] == 'Sunday'] 
saturday = mall_data[mall_data['Weekday'] == 'Saturday']

weeknd = pd.concat([sunday, saturday], axis=0, join='inner')
```

### Средняя посещаемость ТЦ в будни
